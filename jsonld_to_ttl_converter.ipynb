```python
import rdflib

def generate_ttl(jsonld_file, output_ttl):
    g = rdflib.Graph()
    g.parse(jsonld_file, format='json-ld')

    # Add data properties to the graph
    data_properties = [
        "schema",
        "city",
        "coopcycle_url",
        "country",
        "facebook_url",
        "latitude",
        "longitude",
        "mail",
        "name",
        "en",
        "es",
        "fr",
        "eu",
        "url",
        "delivery_form_url",
        "instagram_url",
        "logo_src",
        "twitter_url",
        "faceboook_url"
    ]

    datatypes={
        "schema": "http://schema.org/Text",
        "city": "http://schema.org/Text",
        "coopcycle_url": "http://schema.org/URL",
        "country": "http://schema.org/Text",
        "facebook_url": "http://schema.org/URL",
        "latitude": "http://schema.org/Float",
        "longitude": "http://schema.org/Float",
        "mail": "http://schema.org/Text",
        "name": "http://schema.org/Text",
        "en": "http://schema.org/Text",
        "es": "http://schema.org/Text",
        "fr": "http://schema.org/Text",
        "eu": "http://schema.org/Text",
        "url": "http://schema.org/URL",
        "delivery_form_url": "http://schema.org/URL",
        "instagram_url": "http://schema.org/URL",
        "logo_src": "http://schema.org/URL",
        "twitter_url": "http://schema.org/URL",
        "faceboook_url": "http://schema.org/URL"
    }

    for prop in data_properties:
        # Add the property to the graph with correspondig types (string, float, etc.) according to the datatypes dictionary
        # g.add((rdflib.URIRef("http://schema.org/" + prop), rdflib.RDF.type, rdflib.RDFS.Datatype))
        # add the property without specifying the type
        g.add((rdflib.URIRef("http://schema.org/" + prop), rdflib.RDF.type, rdflib.OWL.DatatypeProperty))

    # Save the Turtle data to a file
    with open(output_ttl, 'w') as f:
        f.write(g.serialize(format='turtle'))

# Usage example
generate_ttl('coopcycle.jsonld', 'converted_data.ttl')
```
